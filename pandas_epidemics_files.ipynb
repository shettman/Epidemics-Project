```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "Resources/hepatitis.csv"
file_two = "Resources/measles.csv"
file_three = "Resources/mumps.csv"
file_four = "Resources/pertussis.csv"
file_five = "Resources/polio.csv"
file_six = "Resources/rubella.csv"
file_seven = "Resources/smallpox.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
file_three_df = pd.read_csv(file_three, encoding="ISO-8859-1")
file_four_df = pd.read_csv(file_four, encoding="ISO-8859-1")
file_five_df = pd.read_csv(file_five, encoding="ISO-8859-1")
file_six_df = pd.read_csv(file_six, encoding="ISO-8859-1")
file_seven_df = pd.read_csv(file_seven, encoding="ISO-8859-1")
```

```python
file_five_df.dropna(subset=["cases"], axis = 0 , inplace= True)
file_five_df.head()
```

```python
# Show just the header
file_one_df.head()
```

```python
renamefile_one_df = file_one_df.rename(columns={"incidence_per_capita": "Hepatitis A incidence_per_capita", "cases": "Hepatitis A cases"})
renamefile_two_df = file_two_df.rename(columns={"incidence_per_capita": "measles incidence_per_capita", "cases": "measles cases"})
renamefile_three_df = file_three_df.rename(columns={"incidence_per_capita": "mumps incidence_per_capita", "cases": "mumps cases" })
renamefile_four_df = file_four_df.rename(columns={"incidence_per_capita": "pertussis incidence_per_capita", "cases": "pertussis cases"})
renamefile_five_df = file_five_df.rename(columns={"incidence_per_capita": "polio incidence_per_capita", "cases": "polio cases"})
renamefile_six_df = file_six_df.rename(columns={"incidence_per_capita": "rubella incidence_per_capita", "cases": "rubella cases"})
renamefile_seven_df = file_seven_df.rename(columns={"incidence_per_capita": "smallpox incidence_per_capita", "cases": "smallpox cases"})

renamefile_five_df.head()
```

```python
#Drop state column
df1 = renamefile_one_df.drop(columns=["state"])
df2 = renamefile_two_df .drop(columns=["state"])
df3 = renamefile_three_df .drop(columns=["state"])
df4 = renamefile_four_df .drop(columns=["state"])
df5 = renamefile_five_df .drop(columns=["state"])
df6 = renamefile_six_df .drop(columns=["state"])
df7 = renamefile_seven_df .drop(columns=["state"])
df1.head()
```

```python
# Drop state_name column
df_1 = df1
df_2 = df2.drop(columns=["state_name"])
df_3 = df3.drop(columns=["state_name"])
df_4 = df4.drop(columns=["state_name"])
df_5 = df5.drop(columns=["state_name"])
df_6 = df6.drop(columns=["state_name"])
df_7 = df7.drop(columns=["state_name"])
df_7.head()
```

```python
# Concat
df_8 = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7], axis=1, join='inner').sort_index()

# Drop week column
df_9 = df_8.drop(columns=["week"])
df_9

```

```python
resetcasesstatedf = df_9.reset_index()
resetcasesstatedf.head()
```

```python
resetcasesstatedf.dtypes
```

```python
groupstatedf = resetcasesstatedf.groupby(["state_name"])
sum_groupstatedf = groupstatedf.sum()
sum_groupstatedf.head()
```

```python
sum_groupstatedf.dtypes
```

```python
groupstatedf = df_9.groupby(["state_name"])
mean_groupstatedf = groupstatedf.mean()
mean_groupstatedf.head()
```

```python
resetcasesstatedf = mean_groupstatedf.reset_index()
resetcasesstatedf.head()
```

```python
#casesstatedf = resetcasesstatedf[["state_name", "Hepatitis A cases", "measles cases","mumps cases", "pertussis cases",  "rubella cases", "smallpox cases"]]
#casesstatedf.head()
```

```python
incidencepercapitadf = resetcasesstatedf[["state_name", "Hepatitis A incidence_per_capita", "measles incidence_per_capita", "mumps incidence_per_capita", "pertussis incidence_per_capita", "polio incidence_per_capita", "rubella incidence_per_capita", "smallpox incidence_per_capita"]]
incidencepercapitadf.head()
```

```python
stacked = incidencepercapitadf.plot(kind='bar', stacked=True, title ="Mean Incidence Per Capita", figsize=(15, 10), legend=True, fontsize=12, color = ["orange", "green", "yellow", "blue", "red", "purple", "violet"])
stacked.set_xlabel("State Name", fontsize=12)
stacked.set_ylabel("Incidence Per Capita", fontsize=12)
plt.ylim(0, 14)
stacked.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
